```python
import matplotlib.pyplot as plt
import numpy as np

a√±os = ["2011", "2012","2013","2014","2015",
       "2016","2017","2018","2019","2020",]

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400
```

# Personalizacion de la Graficas

```python
# introducir series da datos con color, marcador, etiqueta y estilo

# plot: es para los graficos de la serie de lineas

# en el mismo codigo se agraga nombre de etiqueta
# tambien se puede agragar valor a las lineas de la grafica (puede ser exadecimal o poner el color) o se lo deja por defecto
# se le puede agragar un marcador (indica pto que se grafica)
# se puede moficar las lineas de la grafica con linestyle

plt.plot(a√±os, nivel1, label = "Nivel 1", color = "purple", marker = "<", linestyle = "-")
plt.plot(a√±os, nivel2, label = "Nivel 2", color = "red", marker = "*", linestyle = "--")
plt.plot(a√±os, nivel3, label = "Nivel 3", color = "blue", marker = "^", linestyle = ":")
plt.plot(a√±os, nivel4, label = "Nivel 4", color = "black", marker = ".", linestyle = "-.")
plt.plot(a√±os, nivel5, label = "Nivel 5", color = "green", marker = "+", linestyle = " ")



# activar leyendas

plt.legend()

# titulo de grafica
plt.title("examen de certificacion")

# etiquetas a los ejes
plt.xlabel("A√±os que se ha realizado el examen")
plt.ylabel("puntaje")

# personalizar eje de puntaje de 0 a 1000 en incrementos de 50. (hay ticks mayores que son los ptos ppal de la grafica,
# y ticks menores son ptos mas chicos)

# se indica con y o x el eje
plt.yticks(np.arange(0, 1001, 50)) # argumento, se coloca un rango que puede ser numpy o comun. el rango se coloca entre coma y 
                                   # el incremento despues del rango separado por coma. si se quiere ver el tope del rago ej 1000, se + 1

# activar cuadricula

plt.grid()

# activar marcas menores

plt.minorticks_on() # se activa los ticks menores en la grafica = ptos menores

plt.show()  # despliega la grafica

```

# Graficas de barras verticales (bar)

```python
# bar: grafico de barras

eje_x = np.arange(0, 10)

# para que no aparezcan los datos encima del otro y pisen, agregar un desfasaje (ej: 0.2)
# modificar el ancho de barra con width. El ancho por default es de 1 punto

plt.bar(eje_x, nivel5, width = 1/5)
plt.bar(eje_x+0.2, nivel4, width = 1/5)
plt.bar(eje_x+0.4, nivel3, width = 1/5)
plt.bar(eje_x+0.6, nivel2, width = 1/5)
plt.bar(eje_x+0.8, nivel1, width = 1/5)
plt.show()
```

# Grafica de barras horizontales (barh)

```python
plt.barh(eje_x+0.8, nivel5, height = 1/5)
plt.barh(eje_x+0.6, nivel4, height = 1/5)
plt.barh(eje_x+0.4, nivel3, height = 1/5)
plt.barh(eje_x+0.2, nivel2, height = 1/5)
plt.barh(eje_x, nivel1, height = 1/5)
plt.show()
```

# Grafica de barras verticales apiladas

```python
# parametro bottom para apilar las barras o datos
# se muestra los datos apliados no encimados

plt.bar(eje_x, nivel5,)
plt.bar(eje_x, nivel4, bottom = nivel5)
plt.bar(eje_x, nivel3, bottom = nivel4+ nivel5)
plt.bar(eje_x, nivel2, bottom = nivel3+ nivel4+ nivel5)
plt.bar(eje_x, nivel1, bottom = nivel2+ nivel3+ nivel4+ nivel5)
plt.show()
```

# Grafica de dispersion (scatter)

```python
# se le puede agragar marcadores (marker) para diferenciar

plt.scatter(eje_x, nivel1, marker = "<")
plt.scatter(eje_x, nivel2, marker = "*")
plt.scatter(eje_x, nivel3)
plt.scatter(eje_x, nivel4, marker = ".")
plt.scatter(eje_x, nivel5)
plt.show()
```

# Grafica de pie

```python
# cada uno de los valos del array, se va asociar a cada uno de los paises

plt.pie(nivel1,
       labels = ["Mexico", "Colombia", "Espa√±a", "Estados Unidos", "Ecuador", 
                "Argentina", "Venezuela", "Puerto Rico", "Honduras", "Uruguay"])

plt.show()
```

# Combinar graficas

```python
plt.bar(eje_x, nivel1, width = 1/5)
plt.bar(eje_x+0.2, nivel2, width = 1/5)

plt.plot(a√±os, nivel3)

plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel5)

plt.show()
```
